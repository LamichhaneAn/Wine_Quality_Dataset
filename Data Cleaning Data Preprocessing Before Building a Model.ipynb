```python

for i in ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "pH", "sulphates"]:
    df[i].fillna(df[i].median(), inplace = True)

df.isnull().sum()

```

```python
# -  Outliers Treatments:
# Z-score Method: - first find the limits:

upper_limit = df["residual sugar"].mean() + 3*df["residual sugar"].std()
lower_limit = df["residual sugar"].mean() - 3*df["residual sugar"].std()
print("upper_limit : ", upper_limit)
print("lower_limit : ", lower_limit)
```

```python
#find the outliers:

outliers = df[(df["residual sugar"] > upper_limit) | (df["residual sugar"] < lower_limit)]
outliers


```

```python
outliers.shape
```

```python
#triming - delete the outlier data: non_outliers means new_df

non_outliers = df[(df["residual sugar"] < upper_limit) & (df["residual sugar"] > lower_limit)]
non_outliers
```

```python
#triming - delete the outlier data: non_outliers means new_df

non_outliers = df[(df["residual sugar"] < upper_limit) & (df["residual sugar"] > lower_limit)]

print("before removing outliers :", len(df))
print("after removing outliers (non_outliners):", len(non_outliers))
print("Total outliers are : ", len(df) - len(non_outliers))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a boxplot to visualize outliers in the 'residual sugar' column
sns.boxplot(x=non_outliers["residual sugar"])
plt.show()
```

```python
#if you don't want to delete the whole row than we can do capping method
# -  capping - change the outlier values to upper (or) lower limit values:
#it is also filtering the outliers


non_outliers = df.copy()

# Cap values greater than upper_limit
non_outliers.loc[non_outliers["residual sugar"] > upper_limit, "residual sugar"] = upper_limit

# Cap values less than lower_limit
non_outliers.loc[non_outliers["residual sugar"] < lower_limit, "residual sugar"] = lower_limit
```

```python
#Inter Quartile Range Method: IQR

q1 = df["residual sugar"].quantile(0.25)
q3 = df["residual sugar"].quantile(0.75)
iqr = q3-q1

print("q1 :", q1)
print("q3 :", q3)
print("iqr:", iqr)
```

```python
#Next let us calculate the upper and lower limit using the Interquartile Range (IQR) method

upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)

lower_limit , upper_limit
```
